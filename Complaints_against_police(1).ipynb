PLOTTING THE OUTLIERS:

    WE WILL PLOT THE DATAS IN THE BOXPLOT TO KNOW THE INSTANCE OF THE OUTLIERS PRESENT IN THEM.
    THEN WE TRY TO REMOVE THE OUTLIERS USING THE Z-SCORE METHOD.

```python
collist=dfn.columns.values
ncol=11
nrow=11
plt.figure(figsize=(25,25))
for i in range(0,len(collist)):
    plt.subplot(nrow,ncol,i+1)
    sn.boxplot(dfn[collist[i]])
```

```python
from scipy.stats import zscore
z=np.abs(zscore(dfn))
z
```

```python
z[20][15]
```

```python
threshold=3
df_new=dfn[(z<3).all(axis=1)]
df_new
```

```python
import joblib
joblib.dump(df_new,"Complaints_against_police.csv")
```
